```python
cgg_header = pd.read_csv(data_dir + filename, delimiter='\s+', skiprows=1, nrows=46, header=None)
cgg_header
```

```python
x0 = -322.394653613 *1000
y0 = 17559.125629*1000
           #4.4  (rotation)
z0 = 0.5 * 100
```

```python
xvals = cgg_header.values.flatten()[0:110]
yvals = cgg_header.values.flatten()[110:400]
zvals = cgg_header.values.flatten()[400:458]
print(zvals)
```

```python
print(xvals.shape,yvals.shape,zvals.shape)
```

```python
xrng = np.zeros((xvals.shape[0]))
yrng = np.zeros((yvals.shape[0]))
zrng = np.zeros((zvals.shape[0]))

xstart = x0
ystart = y0
zstart = z0

for index, i in enumerate(xvals):
    xrng[index] = xstart
    xstart = xstart + i
    

for index, i in enumerate(yvals):
    
    yrng[index] = ystart
    ystart = ystart + i

for index, i in enumerate(zvals):
    
    zrng[index] = zstart
    zstart = zstart + i
    
```

```python
grid = pv.RectilinearGrid(xrng, yrng, zrng)
```

```python
grid.plot()
```

```python
grid['resistivity'] = cgg_array.flatten(order='F')
```
